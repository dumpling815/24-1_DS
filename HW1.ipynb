```python
q3_data = [13, 16, 20, 20, 20, 21, 25, 25, 29, 30, 33, 35, 35, 35, 40, 52, 71]
mean = sum(q3_data) / len(q3_data)
mean
```

```python
# q3_data[q3_data.index(30)+1:-1]
weighted_trimmed_mean = (sum(q3_data[1:q3_data.index(30)+1]) + 2 * sum(q3_data[q3_data.index(30)+1:-1])) / (len(q3_data[1:q3_data.index(30)+1])+2*len(q3_data[q3_data.index(30)+1:-1]))
weighted_trimmed_mean
```

```python
import matplotlib.pyplot as plt
import numpy as np
q3_data = [13, 16, 20, 20, 20, 21, 25, 25, 29, 30, 33, 35, 35, 35, 40, 52, 71]
```
